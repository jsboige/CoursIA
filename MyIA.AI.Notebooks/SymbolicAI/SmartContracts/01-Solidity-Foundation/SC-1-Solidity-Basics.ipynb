## 1. Structure d'un contrat

```solidity
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.30;

contract MonContrat {
    // Variables d'etat
    // Fonctions
}
```

### Elements

| Element | Description |
|---------|-------------|
| `pragma` | Version du compilateur |
| `contract` | Mot-cle de definition |
| `{ }` | Corps du contrat |

````python
# Exemple de contrat minimal
CONTRACT_EXAMPLE = '''
// SPDX-License-Identifier: MIT
pragma solidity ^0.8.30;

contract Counter {
    uint256 public count;

    function increment() public {
        count += 1;
    }

    function getCount() public view returns (uint256) {
        return count;
    }
}
'''

print("Contrat Counter.sol:")
print(CONTRACT_EXAMPLE)
````

## 2. Types de donnees (Value Types)

### 2.1 Booleens

````python
# Types booleens
BOOL_EXAMPLE = '''
contract BoolExample {
    bool public isActive = true;
    bool public isPaused = false;

    function toggle() public {
        isActive = !isActive;
    }

    function and(bool a, bool b) public pure returns (bool) {
        return a && b;
    }

    function or(bool a, bool b) public pure returns (bool) {
        return a || b;
    }
}
'''

print("Operateurs booleens: && (and), || (or), ! (not)")
print(BOOL_EXAMPLE)
````

### 2.2 Entiers

````python
# Types entiers
INT_EXAMPLE = '''
contract IntExample {
    // Entiers non signes (positifs)
    uint8 public smallNumber;      // 0 a 255
    uint256 public bigNumber;      // 0 a 2^256-1
    uint public defaultUint;       // uint256 par defaut

    // Entiers signes (positifs et negatifs)
    int8 public signedSmall;       // -128 a 127
    int256 public signedBig;       // -2^255 a 2^255-1

    function add(uint a, uint b) public pure returns (uint) {
        return a + b;
    }

    function subtract(int a, int b) public pure returns (int) {
        return a - b;
    }
}
'''

print("Types entiers: uint8, uint256, int8, int256")
print(INT_EXAMPLE)
````

### 2.3 Addresses

````python
# Types adresse
ADDRESS_EXAMPLE = '''
contract AddressExample {
    address public owner;
    address payable public treasury;

    constructor() {
        owner = msg.sender;
    }

    function getBalance(address addr) public view returns (uint256) {
        return addr.balance;
    }

    function sendEther(address payable to) public payable {
        to.transfer(msg.value);
    }
}
'''

print("Types adresse: address, address payable")
print("Proprietes: .balance, .code")
print(ADDRESS_EXAMPLE)
````

### 2.4 Bytes et String

````python
# Types bytes et string
BYTES_EXAMPLE = '''
contract BytesExample {
    bytes32 public fixedData = "hello";
    bytes public dynamicData = "world";
    string public message = "Hello Solidity";

    function setFixed(bytes32 _data) public {
        fixedData = _data;
    }

    function setMessage(string memory _msg) public {
        message = _msg;
    }

    function getLength() public view returns (uint) {
        return bytes(message).length;
    }
}
'''

print("Types bytes: bytes1 a bytes32 (fixes), bytes (dynamique)")
print("Type string: UTF-8, dynamique")
print(BYTES_EXAMPLE)
````

## 3. Variables

### 3.1 Variables d'etat

````python
# Variables d'etat
STATE_VARS = '''
contract StateVariables {
    // Stockees en permanence sur la blockchain
    uint256 public storedData = 42;
    address public owner;
    bool private initialized = false;

    // Visibilite:
    // public - accessible depuis l'exterieur
    // private - accessible uniquement dans ce contrat
    // internal - accessible dans ce contrat et les contrats enfants
    // external - accessible uniquement depuis l'exterieur
}
'''

print("Variables d'etat = stockees sur la blockchain")
print(STATE_VARS)
````

### 3.2 Variables locales

````python
# Variables locales
LOCAL_VARS = '''
contract LocalVariables {
    uint256 public result;

    function calculate(uint a, uint b) public returns (uint256) {
        // Variables locales (non stockees)
        uint256 sum = a + b;
        uint256 product = a * b;

        // Stockage du resultat
        result = sum + product;
        return result;
    }
}
'''

print("Variables locales = temporaires, non stockees")
print(LOCAL_VARS)
````

### 3.3 Variables globales

````python
# Variables globales
GLOBAL_VARS = '''
contract GlobalVariables {
    function getGlobals() public view returns (
        address sender,
        uint256 value,
        uint256 timestamp,
        uint256 blockNumber,
        uint256 gasLimit
    ) {
        return (
            msg.sender,     // Adresse de l'appelant
            msg.value,      // ETH envoye (en wei)
            block.timestamp,// Horodatage du bloc
            block.number,   // Numero du bloc
            block.gaslimit  // Limite de gaz du bloc
        );
    }
}
'''

print("Variables globales Solidity:")
print("  msg.sender - Adresse de l'appelant")
print("  msg.value - ETH envoye")
print("  msg.data - Donnees de l'appel")
print("  block.timestamp - Horodatage")
print("  block.number - Numero du bloc")
print("  tx.origin - Createur de la transaction")
print(GLOBAL_VARS)
````

## 4. Conversions de types

````python
# Conversions
CONVERSION_EXAMPLE = '''
contract TypeConversions {
    // Conversion explicite
    function uintToInt(uint256 a) public pure returns (int256) {
        return int256(a);
    }

    // Conversion address to payable
    function toPayable(address addr) public pure returns (address payable) {
        return payable(addr);
    }

    // String to bytes
    function stringToBytes(string memory s) public pure returns (bytes memory) {
        return bytes(s);
    }

    // Address to uint (pour comparaison)
    function addressToUint(address addr) public pure returns (uint160) {
        return uint160(addr);
    }
}
'''

print("Conversions de types Solidity:")
print(CONVERSION_EXAMPLE)
````
