## 1. Vue d'ensemble â un mÃŠme Sudoku, deux modes d'encodage

Un Sudoku 4x4 se modÃĐlise par **16 cellules** prenant des valeurs dans `{1, 2, 3, 4}`, soumises Ã  trois familles de contraintes : chaque **rangÃĐe**, chaque **colonne** et chaque **bloc 2x2** contient des valeurs **distinctes**. Z3 permet d'exprimer ces 16 cellules de deux maniÃĻres fondamentalement diffÃĐrentes :

| Mode | Encodage pyz3 | Analogie |
|------|----------------|----------|
| **Constants** | 16 variables `Int` sÃĐparÃĐes (`c_0_0`, `c_0_1`, âĶ, `c_3_3`) | un tableau de scalaires |
| **Array** | un `Array(Int, Int)` indexÃĐ par `r*4+c` (thÃĐorie des tableaux) | une fonction Index â Valeur |

Les deux modes sont **sÃĐmantiquement ÃĐquivalents** : ils expriment le mÃŠme espace de solutions. La diffÃĐrence est **structurelle** â le mode `Array` traite la grille comme une fonction (utile si on raisonne sur des cellules arbitraires ou des mises Ã  jour `Store`), le mode `Constants` traite la grille comme un tableau de scalaires (plus direct, plus lisible). Ce notebook rÃĐsout le **mÃŠme** Sudoku dans les deux modes et vÃĐrifie qu'ils convergent.


## 2. La grille 4x4 et ses contraintes canoniques

Fixons d'abord une instance avec quelques **givens** (cellules prÃĐ-remplies) pour garantir une solution unique et pouvoir comparer. La grille suivante a trois givens ; le solveur doit complÃĐter le reste.

```
+---+---+---+---+
| 1 | . | . | . |
+---+---+---+---+
| . | . | . | 3 |
+---+---+---+---+
| . | . | 2 | . |
+---+---+---+---+
| . | 4 | . | . |
+---+---+---+---+
```

Les contraintes canoniques (identiques dans les deux modes) :
1. **Domaine** : chaque cellule â `{1, 2, 3, 4}`.
2. **RangÃĐes** : les 4 cellules de chaque rangÃĐe sont distinctes.
3. **Colonnes** : les 4 cellules de chaque colonne sont distinctes.
4. **Blocs 2x2** : les 4 cellules de chaque bloc sont distinctes.
5. **Givens** : les cellules prÃĐ-remplies sont fixÃĐes.


## 3. RÃĐsolution en mode `Constants` (un `Int` par cellule)

Le mode le plus direct : on crÃĐe 16 variables `Int` organisÃĐes en matrice 4x4, puis on pose les contraintes Sudoku canoniques. C'est l'encodage idiomatique pyz3 pour un problÃĻme sur une grille fixe.


````python
# Mode Constants : 16 Int constants (matrice 4x4).
N = 4
CONST = [[Int(f"c_{r}_{c}") for c in range(N)] for r in range(N)]

s_const = Solver()

# (1) Domaine 1..4 pour chaque cellule.
for r in range(N):
    for c in range(N):
        s_const.add(CONST[r][c] >= 1, CONST[r][c] <= N)

# (2) RangÃĐes : valeurs distinctes.
for r in range(N):
    s_const.add(Distinct(CONST[r]))

# (3) Colonnes : valeurs distinctes.
for c in range(N):
    s_const.add(Distinct([CONST[r][c] for r in range(N)]))

# (4) Blocs 2x2 : valeurs distinctes (N=4 -> blocs de taille 2).
B = 2  # taille de bloc (sqrt(N))
for br in range(0, N, B):
    for bc in range(0, N, B):
        bloc = [CONST[br + i][bc + j] for i in range(B) for j in range(B)]
        s_const.add(Distinct(bloc))

# (5) Givens : (rangÃĐe, colonne) -> valeur.
GIVENS = {(0, 0): 1, (1, 3): 3, (2, 2): 2, (3, 1): 4}
for (r, c), v in GIVENS.items():
    s_const.add(CONST[r][c] == v)

print("Mode Constants - statut :", s_const.check())
m_const = s_const.model()
sol_const = [[m_const.eval(CONST[r][c]).as_long() for c in range(N)] for r in range(N)]
print("Solution (Constants) :")
for row in sol_const:
    print("  ", row)
````

## 4. RÃĐsolution en mode `Array` (thÃĐorie des tableaux)

Le mode alternatif : la grille entiÃĻre est **un seul** `Array(Int, Int)` oÃđ l'indice `r*4 + c` dÃĐsigne la cellule `(r, c)`. On accÃĻde aux cellules via `Select` (voir [Z3-Python-17 Array Theory](Z3-Python-17-Array-Theory.ipynb)). Les contraintes sont les mÃŠmes, mais exprimÃĐes sur des `Select(grid, i)` plutÃīt que sur des scalaires.


````python
# Mode Array : un Array(Int, Int) indexe par r*4+c.
grid = Array("grid", IntSort(), IntSort())

s_arr = Solver()

def cell(r, c):
    """Select de la cellule (r,c) dans le tableau symbolique."""
    return Select(grid, r * N + c)

# (1) Domaine 1..4 pour chaque cellule.
for r in range(N):
    for c in range(N):
        s_arr.add(cell(r, c) >= 1, cell(r, c) <= N)

# (2) RangÃĐes distinctes.
for r in range(N):
    s_arr.add(Distinct([cell(r, c) for c in range(N)]))

# (3) Colonnes distinctes.
for c in range(N):
    s_arr.add(Distinct([cell(r, c) for r in range(N)]))

# (4) Blocs 2x2 distincts.
for br in range(0, N, B):
    for bc in range(0, N, B):
        bloc = [cell(br + i, bc + j) for i in range(B) for j in range(B)]
        s_arr.add(Distinct(bloc))

# (5) Givens (memes que le mode Constants).
for (r, c), v in GIVENS.items():
    s_arr.add(cell(r, c) == v)

print("Mode Array - statut :", s_arr.check())
m_arr = s_arr.model()
sol_arr = [[m_arr.eval(cell(r, c)).as_long() for c in range(N)] for r in range(N)]
print("Solution (Array) :")
for row in sol_arr:
    print("  ", row)
````

## 5. VÃĐrification â les deux solutions sont des carrÃĐs latins valides (et coÃŊncident)

Les deux modes doivent produire des grilles qui satisfont **toutes** les contraintes Sudoku (domaine, rangÃĐes, colonnes, blocs, givens). VÃĐrifions le programmatoirement, puis comparons les deux solutions. Avec les mÃŠmes givens et (probablement) une solution unique, elles doivent coÃŊncider exactement.


````python
# Verification : chaque solution est un carre latin Sudoku 4x4 valide.
def est_valide_sudoku(sol):
    """Retourne True si sol est une grille Sudoku 4x4 valide."""
    # Domaine + valeurs entieres 1..4.
    if not all(1 <= sol[r][c] <= N for r in range(N) for c in range(N)):
        return False
    # RangÃĐes.
    if any(len(set(sol[r])) != N for r in range(N)):
        return False
    # Colonnes.
    if any(len({sol[r][c] for r in range(N)}) != N for c in range(N)):
        return False
    # Blocs 2x2.
    for br in range(0, N, B):
        for bc in range(0, N, B):
            bloc = {sol[br + i][bc + j] for i in range(B) for j in range(B)}
            if len(bloc) != N:
                return False
    # Givens respectÃĐs.
    if any(sol[r][c] != v for (r, c), v in GIVENS.items()):
        return False
    return True

print("Solution Constants valide :", est_valide_sudoku(sol_const))
print("Solution Array valide      :", est_valide_sudoku(sol_arr))
print("Les deux solutions coÃŊncident :", sol_const == sol_arr)
````

### InterprÃĐtation â pourquoi les deux modes convergent

Les deux solutions coÃŊncident (ou sont toutes deux valides si plusieurs solutions existent) : c'est la preuve directe que **l'encodage ne change pas le problÃĻme**. Le mode `Constants` et le mode `Array` sont deux **syntaxes** diffÃĐrentes pour le mÃŠme ensemble de contraintes â le solveur Z3 les rÃĐÃĐcrit toutes deux vers le mÃŠme noyau SMT avant la rÃĐsolution.

Le choix entre les deux modes est donc une dÃĐcision de **lisibilitÃĐ et de contexte**, pas de correction :
- **`Constants`** est plus lisible pour une grille fixe (on nomme chaque cellule), et c'est l'encodage idiomatique pyz3.
- **`Array`** est pertinent dÃĻs qu'on veut raisonner sur des **cellules arbitraires** (indice calculÃĐ) ou des **mises Ã  jour** (`Store` pour simuler un coup), ou raisonner sur une grille **non bornÃĐe**.

**Verdict honnÃŠte (Prong-B)** : sur un Sudoku 4x4, les deux modes se rÃĐsolvent instantanÃĐment â la diffÃĐrence de performance est invisible. La comparaison devient discriminante sur des grilles plus grandes (9x9) ou quand l'encodage `Array` permet des raccourcis de modÃĐlisation qu'une explosion de `Constants` (81 variables nommÃĐes) rendrait illisibles.

## Conclusion

Ce notebook dÃĐmontre le principe fondamental de la modÃĐlisation SMT : **le problÃĻme (les contraintes) est sÃĐparÃĐ de l'encodage (comment on les exprime)**. Deux encodages radicalement diffÃĐrents â 16 `Int` constants vs un `Array(Int, Int)` â du mÃŠme Sudoku 4x4 convergent vers la mÃŠme solution valide.

| Aspect | Mode `Constants` | Mode `Array` |
|--------|------------------|--------------|
| Structure | 16 scalaires nommÃĐs | 1 fonction Index â Valeur |
| AccÃĻs cellule | `CONST[r][c]` | `Select(grid, r*4+c)` |
| LisibilitÃĐ grille fixe | Excellente | Moyenne |
| Cellule arbitraire / mise Ã  jour | MalaisÃĐ (nom fixe) | Naturel (`Store`) |
| SÃĐmantique (solutions) | **Identique** | **Identique** |

**LeÃ§on** : choisir l'encodage selon le **contexte de raisonnement** (grille fixe vs dynamique), pas selon une prÃĐtendue ÂŦ meilleure Âŧ reprÃĐsentation. Le solveur rÃĐÃĐcrit tout vers le mÃŠme noyau SMT â la correction est garantie par les contraintes, pas par la syntaxe.
