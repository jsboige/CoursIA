## 1. Biographie et style singulier

Ne a Liverpool en 1937, John Conway etudie et enseigne a **Cambridge** (Gonville & Caius College)
de 1964 a 1986, puis a **Princeton** jusqu'a sa mort en avril 2020 (du COVID-19).

Son style mathematique est unique : **le jeu comme methode**. La ou d'autres voient un divertissement,
Conway voit une structure profonde a formaliser. Le Jeu de la Vie n'est pas une recreation : c'est une
machine de Turing universelle deguisee. Les jeux a deux joueurs (Nim, Hackenbush, Domineering) ne sont
pas des passe-temps : ils engendrent une **arithmetique** complete — les nombres surreels.

Cette **espieglerie au service de la profondeur** caracterise toute son oeuvre. Conway aimait
inventer des notations frappantes (la fleche de Knuth-Conway, la notation des groupes), des noms
imagees (le "Monstre", les "lutins" / *sprouts*), et des algorithmes que l'on peut executer de tete
(l'algorithme Doomsday pour calculer le jour de la semaine).

Trois livres jalonnent son oeuvre accessible :

- ***On Numbers and Games*** (1976) — les nombres surreels et la theorie des jeux combinatoires.
- ***Winning Ways for Your Mathematical Plays*** (1982, avec Berlekamp et Guy) — la bible des jeux combinatoires.
- ***Sphere Packings, Lattices and Groups*** (1988, avec Sloane) — "SPLAG", reseaux et groupes.


## 2. Panorama des grands resultats

Conway a laisse une empreinte dans des domaines extraordinairement varies. En voici un tour
d'horizon — c'est cette **largeur** que l'Epic met en avant, et non le seul Game of Life.


### 2.1 Les nombres surreels

Dans *On Numbers and Games*, Conway construit a partir des jeux combinatoires une classe de nombres,
les **nombres surreels** (note **No**), qui contient a la fois les reels, les ordinaux transfinis et
des infinitesimaux. Chaque nombre est defini recursivement comme une paire $\{L \mid R\}$ d'ensembles
de surreels deja construits, avec la contrainte qu'aucun element de $L$ ne soit $\geq$ a un element de $R$.

C'est le **plus grand corps totalement ordonne** possible (une "classe propre"). On y retrouve $0=\{\mid\}$,
$1=\{0\mid\}$, $\tfrac12=\{0\mid 1\}$, mais aussi $\omega=\{0,1,2,\dots\mid\}$ et $\tfrac1\omega=\{0\mid 1,\tfrac12,\tfrac14,\dots\}$.
Donald Knuth a popularise cette construction dans son roman mathematique *Surreal Numbers* (1974),
ou il forge d'ailleurs le terme "surreal".


### 2.2 Groupes de Conway et Monstrous Moonshine

En etudiant le **reseau de Leech** (cf. 2.3), Conway decouvre en 1968 son groupe d'automorphismes et
en extrait trois **groupes sporadiques** nouveaux : $\mathrm{Co}_1$, $\mathrm{Co}_2$, $\mathrm{Co}_3$
(le groupe complet $\mathrm{Co}_0 = 2\cdot\mathrm{Co}_1$ a pour ordre environ $8{,}3\times10^{18}$).

Ces groupes s'inscrivent dans la classification des **groupes finis simples**, et notamment dans la
constellation du **Monstre** (le plus grand groupe sporadique). La conjecture du **Monstrous Moonshine**,
formulee par Conway et Norton en 1979, relie de maniere stupefiante les dimensions des representations
du Monstre aux coefficients de la fonction modulaire $j$. Elle sera demontree par Richard Borcherds
(medaille Fields 1998).


### 2.3 Reseau de Leech et empilement de spheres

Le **reseau de Leech** $\Lambda_{24}$ est un reseau exceptionnel en dimension 24. Conway en calcule
le groupe d'automorphismes et montre qu'il realise l'empilement de spheres le plus dense connu en
dimension 24, avec un **nombre de contact** (kissing number) de **196 560** — chaque sphere en touche
196 560 autres. La question de l'optimalite de cet empilement en dimension 24 (et en dimension 8 avec
le reseau $E_8$) a finalement ete tranchee en 2016 par Maryna Viazovska, prolongeant directement
l'heritage de Conway.


### 2.4 Polynome de Conway (theorie des noeuds)

Conway reformule le polynome d'Alexander d'un noeud sous une forme calculable par une **relation
d'echeveau** (skein relation) locale, donnant le **polynome d'Alexander-Conway** $\nabla(z)$ :

$$\nabla(L_+) - \nabla(L_-) = z\,\nabla(L_0)$$

ou $L_+, L_-, L_0$ sont trois diagrammes identiques sauf au voisinage d'un croisement. Cette approche
combinatoire et locale a ouvert la voie aux invariants modernes (polynome de Jones, HOMFLY).


### 2.5 Algorithme Doomsday

L'**algorithme Doomsday** (1973) permet de calculer **de tete** le jour de la semaine de n'importe
quelle date. L'idee : dans chaque annee, certaines dates "ancres" (4/4, 6/6, 8/8, 10/10, 12/12, le
dernier jour de fevrier...) tombent toutes le meme jour de la semaine, le *Doomsday* de l'annee. Il
suffit alors de se deplacer modulo 7. C'est l'archetype de la mathematique "jouable" de Conway.
**On en formalise une version en Lean (section 3).**


### 2.6 Constante de Look-and-Say

La suite **"look-and-say"** se lit a voix haute : $1, 11, 21, 1211, 111221, 312211, \dots$ (on decrit
le terme precedent : "un 1" -> 11, "deux 1" -> 21, etc.). Conway demontre le **theoreme cosmologique** :
toute chaine se decompose a terme en 92 "elements" atomiques. Surtout, le rapport des longueurs de
termes consecutifs converge vers la **constante de Conway** $\lambda \approx 1{,}303577269\dots$,
unique racine reelle positive d'un polynome de degre 71. **On en formalise des cas en Lean (section 3).**


### 2.7 FRACTRAN

**FRACTRAN** est un langage de programmation esoterique invente par Conway : un programme est une
**liste de fractions**, et l'execution multiplie un entier courant par la premiere fraction qui donne
un entier, jusqu'a ce qu'aucune ne convienne. Aussi minimaliste soit-il, FRACTRAN est **Turing-complet** ;
le programme PRIMEGAME de Conway engendre la suite des nombres premiers dans les exposants de 2.


### 2.8 Probleme de l'Ange

Le **probleme de l'Ange** (Conway, 1982) oppose un *Ange* de puissance $k$ (qui saute jusqu'a $k$ cases
en distance de Chebyshev) a un *Diable* qui detruit une case par tour. L'Ange peut-il echapper
indefiniment ? Conway met une prime de 1000 $ sur la question. Elle est resolue **en 2006** par quatre
auteurs independants (Kloster, Mathe, Bowditch, Gacs) : un Ange de puissance $\geq 2$ gagne toujours.
**On formalise la combinatoire du moveset en Lean (section 3).**


### 2.9 Sprouts et le libre arbitre

- **Sprouts** : un jeu topologique de papier-crayon invente par Conway et Michael Paterson (1967),
  ou l'on relie des points par des courbes sans croisement. Sa theorie combinatoire est etonnamment riche.
- **Theoreme du libre arbitre** (Conway-Kochen, 2006) : si les experimentateurs ont un "libre arbitre"
  (leurs choix de mesure ne sont pas determines par le passe), alors certaines particules elementaires
  ont la meme propriete. Ce resultat s'appuie sur le **theoreme de Kochen-Specker** — c'est l'objet
  du notebook [Lean-15](Lean-15-Kochen-Specker.ipynb) **(NB3 de l'Epic)**.

Le pattern d'execution repose sur **WSL + Lake** (cf. [Lean-14](Lean-14-Conway-Tribute.ipynb) et
[Lean-12](Lean-12-Sensitivity-Theorem.ipynb)). La **detection de chemin** rend le notebook portable :

```python
# Setup : detection de chemin portable + helper WSL/Lake (single source of truth)
import subprocess
import re
from pathlib import Path

def find_conway_lean_project():
    """Auto-detecte conway_lean/ a partir de l'emplacement du notebook (sans hardcode)."""
    current = Path.cwd()
    for _ in range(6):
        candidate = current / 'conway_lean'
        if candidate.exists() and (candidate / 'lakefile.lean').exists():
            return candidate
        lean_dir = current / 'MyIA.AI.Notebooks' / 'SymbolicAI' / 'Lean'
        candidate = lean_dir / 'conway_lean'
        if candidate.exists() and (candidate / 'lakefile.lean').exists():
            return candidate
        current = current.parent
    raise FileNotFoundError("conway_lean/ introuvable - verifier le repertoire de travail")

def win_to_wsl(win_path: Path) -> str:
    """Convertit un chemin Windows en chemin WSL via la lettre de lecteur (portable c:/d:/...)."""
    drive = win_path.drive[0].lower()
    return f'/mnt/{drive}{win_path.as_posix()[2:]}'

def wsl(cmd, timeout=60):
    """Execute une commande bash dans WSL Ubuntu."""
    full = ['wsl', '-d', 'Ubuntu', '--', 'bash', '-lc', cmd]
    try:
        r = subprocess.run(full, capture_output=True, text=True, timeout=timeout)
        return r.returncode, r.stdout, r.stderr
    except subprocess.TimeoutExpired:
        return -1, '', f'TIMEOUT apres {timeout}s'
    except FileNotFoundError:
        return -2, '', 'WSL introuvable (notebook concu pour Windows + WSL Ubuntu)'

WIN_LEAN_PROJECT = find_conway_lean_project()
LEAN_PROJECT = win_to_wsl(WIN_LEAN_PROJECT)

print(f"Projet Lean (Windows) : {WIN_LEAN_PROJECT}")
print(f"Projet Lean (WSL)     : {LEAN_PROJECT}")
print(f"Toolchain             : {(WIN_LEAN_PROJECT / 'lean-toolchain').read_text(encoding='utf-8').strip()}")
assert (WIN_LEAN_PROJECT / 'lakefile.lean').exists(), 'conway_lean/lakefile.lean introuvable'
```

```python
# Helpers de lecture single-source-of-truth des .lean (lecture directe du depot)
DECL_RE = re.compile(r"^\s*(theorem|lemma|def|abbrev|instance)\s+([\w'À-ſ]+)")

def count_real_sorry(text: str) -> int:
    """Compte les vrais `sorry` tactiques en excluant commentaires de ligne et blocs /- -/."""
    n, in_block = 0, False
    for line in text.splitlines():
        s = line.strip()
        if in_block:
            if '-/' in s:
                in_block = False
            continue
        if s.startswith('/-'):
            if '-/' not in s:
                in_block = True
            continue
        code_part = s.split('--', 1)[0]
        if re.search(r'\bsorry\b', code_part):
            n += 1
    return n

def show_lean(relpath: str, max_lines: int = 0):
    """Affiche les declarations (theorem/def/...) d'un fichier .lean reel + comptage sorry."""
    f = WIN_LEAN_PROJECT / relpath
    text = f.read_text(encoding='utf-8')
    decls = [(i + 1, m.group(1), m.group(2))
             for i, l in enumerate(text.splitlines())
             for m in [DECL_RE.match(l)] if m]
    sorry = count_real_sorry(text)
    print(f"=== {relpath}  ({len(text.splitlines())} lignes, sorry reels = {sorry}) ===")
    for ln, kind, name in decls:
        print(f"  L{ln:>4}  {kind:<8} {name}")
    if max_lines:
        print("-" * 60)
        for l in text.splitlines()[:max_lines]:
            print(l)
    return sorry
```
