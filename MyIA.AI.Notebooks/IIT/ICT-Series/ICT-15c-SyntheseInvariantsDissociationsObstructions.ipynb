## Objet

Document de **synthèse transversale** (pas un nouveau dispatch expérimental).
La série ICT a produit une batterie de substrats instrumentés et de proxys ; il
émerge de cette moisson **trois régimes de lecture** d'une trajectoire, applicables
à chaque substrat déjà instancié. Cette notebook les pose **opérationnellement**
(pas seulement conceptuellement) et les relie aux deux chantiers actifs (N1 et N2).

## Régime 1 — Invariants

**Définition opérationnelle.** Quantités ou signatures qui **persistent** quand on
change de substrat, de granularité temporelle, ou de proxy. L'invariant est ce qui
se transporte d'un cas à l'autre sans modification — c'est l'objet qu'on espère
voir émerger au terme d'une bonne théorie cross-substrat.

**Question directrice** : *qu'est-ce qui reste constant à travers les substrats ?*

**Tests falsifiables** :
- Reproduire la même valeur d'invariant sur ≥ 3 substrats distincts.
- Le ranger (Kendall tau, Spearman rho) doit rester élevé cross-substrat.
- Disparition de l'invariant sous perturbation controlée = bon test de spécificité.

**Lecture transversale** : les invariants ne sont **pas** un scalaire unique. Ce sont
des *familles* d'invariants, chacune stable sur sa fenêtre de substrats. La réunion
de toutes ces familles n'est pas recolable en un seul nombre — c'est précisément
ce que le régime 3 (obstructions) documente.

## Régime 2 — Dissociations

**Définition opérationnelle.** Cas où **deux proxys censés mesurer la même chose
divergent**, alors qu'on les croyait alignés. La dissociation n'est pas un échec
de mesure mais un **signal** : elle révèle que les deux proxys ne mesuraient pas
la même grandeur (ou la même grandeur sous le même angle).

**Question directrice** : *où les proxys se séparent-ils, et qu'est-ce que la
séparation révèle ?*

**Tests falsifiables** :
- Cross-corrélation / rang (Kendall tau) entre deux proxys significativement < 1.
- Visualisation de la dissociation : heatmap, scatter (proxy_a vs proxy_b).
- Documenter ce que la divergence **révèle** du substrat (pas la masquer).

**Lecture transversale** : la dissociation est **structurellement informative**.
Elle ne disqualifie aucun des deux proxys ; elle identifie leur **portée
respective**. Φ et F mesurent la même grandeur intégrée ; K mesure un troisième
axe (compression) qui n'est pas sous-jacent aux deux premiers — leur corrélation
partielle est *attendue*, pas un défaut.

## Régime 3 — Obstructions

**Définition opérationnelle.** Cas où le recollement **échoue** : on ne peut pas
recoller des sections locales en une section globale cohérente. L'obstruction
n'est pas un échec de méthode mais une **structure** (cohomologie `H¹ ≠ 0`).
La classe d'obstruction porte l'information que les sections locales refusent
de transmettre au global.

**Question directrice** : *qu'est-ce qui empêche le recollement global, et cette
impossibilité est-elle stable cross-substrat ?*

**Tests falsifiables** :
- Le détecteur méta-proxy (ICT-15c, c.752) renvoie-t-il `STABLE` (motif reproductible
  cross-substrat) ou `NOISE` (motif aléatoire) sur les substrats de référence ?
- Test SAT/UNSAT sur les systèmes à contraintes (cf. Kochen-Specker, ICT).
- Production d'entropie σ > 0 sur la réversibilisation d'une trajectoire (cf.
  ICT-18, ICT-18b) : quantité effacée par la projection réversible.

**Lecture transversale** : la stabilité cross-substrat du motif d'obstruction est
l'**objet expérimental** de N1 [#7395](https://github.com/jsboige/CoursIA/issues/7395).
Tant que le détecteur méta-proxy ne renvoie pas `STABLE`, **l'objet n'est pas
encore nommé** (discipline de nommage de #7395) — c'est précisément la valeur
diagnostique du verdict `NOISE` documenté en c.752 : le détecteur borne la
divergence et démontre qu'il ne retourne pas `STABLE` sur n'importe quoi
(anti-faux-positif validé).

## Articulation avec N1 et N2

### N1 — Méta-proxy d'obstruction ([#7395](https://github.com/jsboige/CoursIA/issues/7395))

Le détecteur `ict.meta_proxy` (c.752, PR #7578) **opérationnalise le régime 3** :
il produit le vecteur d'obstruction `(a - b) / (|a| + |b| + eps)` entre signatures
multi-proxys de substrats distincts, et le verdict falsifiable `STABLE / NOISE /
INCONCLUSIVE`. La classe d'obstruction (si elle est stable cross-substrat) est
l'invariant que le régime 3 cherche.

**État post-c.752** : verdict = `NOISE` documenté. L'objet sous-jacent (la classe
d'obstruction) **n'est pas stabilisé empiriquement**. La jambe meurt proprement —
le détecteur est livré, l'objet n'est pas nommé (cf. discipline de nommage de
[#7395](https://github.com/jsboige/CoursIA/issues/7395) : « Ne pas nommer l'objet
tant qu'il n'est pas stabilisé »).

### N2 — Trajectoires de représentations ([#7396](https://github.com/jsboige/CoursIA/issues/7396))

N2 déplace la question des **états** vers les **représentations** : l'objet suivi
n'est plus l'état du système mais sa représentation interne (apprise). Les jambes
déjà esquissées (Grokking / SAE Qwen / Personae-RL / J-Lens / GWT) relèvent toutes
de cette famille. La grille **invariants / dissociations / obstructions** est
directement applicable à ce nouveau plan :

- **Invariants** sur les représentations : la compression progressive (Schmidhuber)
  est-elle invariante cross-représentation ?
- **Dissociations** sur les représentations : SAE features vs J-space vs persona
  pathology — où divergent-elles, et que révèle la divergence ?
- **Obstructions** sur les représentations : l'inoculation (#5105) relève-t-elle
  à dessein l'obstruction à un mode de représentation pathologique ?

**État actuel** : N2 = cadrage stratégique, non-argumentation, **gated** sur
stabilisation de N1. La grille fournit le vocabulaire pour les futures
livrables N2, mais ne présuppose pas leur existence.

## Discipline de nommage (HARD)

L'objet sous-jacent au régime 3 (la classe d'obstruction cross-substrat) **n'est
pas encore nommé**. Le vocabulaire utilisé dans cette synthèse reste descriptif :

- *invariant* (au sens de « ce qui persiste »), pas *canonique* ;
- *dissociation* (au sens de « divergence entre proxys »), pas *paradoxal* ;
- *obstruction* (au sens de « classe de cohomologie H¹ »), pas *barrière* ou
  *limite*.

Aucune lettre grecque décorée sur un objet non encore constaté. La stabilisation
empirique (via N1) précède le baptême. Cette discipline vaut aussi pour N2 :
tant que les jambes (Grokking / SAE / Personae-RL / J-Lens) ne sont pas
recollables, on décrit la famille, on ne force pas une théorie unique.

## Suite logique

Cette synthèse est **statique** (pas de nouveau dispatch expérimental). Sa valeur
tient à la **consolidation** qu'elle opère :

1. Elle fournit un vocabulaire commun pour les futurs notebooks ICT
   (chaque nouveau substrat gagne une lecture dans les trois régimes).
2. Elle relie les chantiers actifs (N1, N2) à un cadre pré-existant sans
   présupposer leur stabilisation.
3. Elle sert de **référentiel** pour le recalibrage éventuel des seuils du
   détecteur méta-proxy (c.752 : seuils `0.05 / 0.30` — à recalibrer si
   recalibrage Gray-Scott F/k ouvre une fenêtre `STABLE`).

Si une re-exécution d'ICT-15c avec Gray-Scott recalibré (e.g. `F=0.046,
k=0.063`, fenêtre Turing plus chaude) renvoie `STABLE`, alors la grille gagne
un cas positif documenté — et l'objet sous-jacent peut être nommé. Tant que
ce recalibrage n'est pas fait, la synthèse est honnête : trois régimes,
chacun illustré par ≥ 5 cas concrets, sans unification prématurée.
