## La thèse (cadrage user 2026-07-18)

La **réversibilisation** comme principe fondationnel unifiant a une clé : la **collaboration agentique**. Observation directrice : la plupart des processus irréversibles — physiques, biologiques ou culturels — pointent vers l'**aliénation de la collaboration**, l'asservissement, ou la **destruction de l'altérité agentique**. Symétriquement, ce que la réversibilité *permet* (résistance à l'entropie, histoire de la vie, culture) passe par le maintien d'agents **autres** capables de coopérer sans être absorbés.

C'est une thèse séduisante — et exactement le genre de thèse qui devient une *just-so story* si on la laisse s'auto-confirmer. D'où ce contrat, fixé **avant** tout run empirique.

## 4. Smoke test — calibration des 3 candidats sur cas synthétiques

Avant tout run sur substrat réel, on vérifie que les 3 candidats **réagissent dans le sens attendu** sur deux cas synthétiques extrêmes : une **monoculture** (agents identiques) et une **coalition** (agents distincts coordonnés). Ce n'est pas un test de la thèse — c'est un test que les *instruments* mesurent bien ce qu'ils doivent mesurer.

```python
# Deux populations synthétiques : monoculture vs coalition
rng = np.random.default_rng(42)

# Monoculture : 6 agents, consignes identiques + petit bruit
setpoints_mono = rng.normal(loc=0.0, scale=0.01, size=(6, 2))
# Coalition : 6 agents en 3 paires de consignes distinctes
centers = np.array([[0., 0.], [5., 0.], [2.5, 4.]])
setpoints_coal = np.vstack([rng.normal(c, 0.1, size=(2, 2)) for c in centers])

print('=== Candidat 1 : setpoint_diversity (pairwise_mean) ===')
print(f'  monoculture : {ao.setpoint_diversity(setpoints_mono):.4f}')
print(f'  coalition   : {ao.setpoint_diversity(setpoints_coal):.4f}  (attendu >>)')

print('=== Candidat 2 : policy_mutual_information ===')
# Monoculture comportementale : A et B toujours identiques (absorption)
jc_abs = np.diag([20., 20., 20.])
# Coordination sans absorption : A et B corrélés mais distincts
jc_coal = np.array([[15., 5., 0.], [5., 15., 0.], [0., 0., 0.]])
# Indépendance (référence basse)
jc_indep = np.outer([20., 20., 20.], [13., 13., 14.])
print(f'  indépendance    : {ao.policy_mutual_information(jc_indep):.4f}  (attendu ~0)')
print(f'  absorption      : {ao.policy_mutual_information(jc_abs):.4f}  (élevée)')
print(f'  coordination    : {ao.policy_mutual_information(jc_coal):.4f}  (intermédiaire)')
print('  => note : MI seule ne distingue pas absorption de coordination (cf. piège §3)')

print('=== Candidat 3 : occupied_basin_count (n_neighbors=5) ===')
traj_mono = rng.normal(0.0, 0.05, size=(30, 2))
traj_coal = np.vstack([rng.normal(c, 0.05, size=(15, 2)) for c in centers])
print(f'  monoculture : {ao.occupied_basin_count(traj_mono)} bassin  (attendu 1)')
print(f'  coalition   : {ao.occupied_basin_count(traj_coal)} bassins (attendu 3)')
```

### P2 — Contre-régime : signature temporelle (clause 3, centrale)
Une monoculture **forcée** présente un budget `B_state` *élevé à court terme* mais à **queue lourde** (effondrement différé sous perturbation corrélée), tandis qu'une coalition présente un budget *modéré mais résilient*. Si `B_state` ne distingue pas les deux régimes par leur signature temporelle, la thèse est **raffinée** (pas affirmée).

### P3 — Triangulation des candidats mesure
Les 3 candidats (consignes, MI, bassins) **convergent** sur le classement monoculture < coalition sur au moins 2 substrats. S'ils divergent, la dissociation est **publiée** (pas masquée par un choix *ad hoc* post-données).
